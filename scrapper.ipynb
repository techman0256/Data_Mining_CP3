```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

### scrapped fixtures of all matches of the season 2023.

```python
page = requests.get('https://www.cricketworldcup.com/fixtures')
# print(page)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
class Match:
    def __init__(self, team1, team2, date, time, venue, conclusion):
        self.team1 = team1
        self.team2 = team2
        self.date = date
        self.time = time
        self.venue = venue
        self.conclusion = conclusion
```

```python
data = []

matches = soup.find_all('div', class_='match-block')
print(len(matches))
# print(matches[0].contents[1].text)
# print(matches[0].contents[3].contents[1])

for match in matches:
    conclusion = match.contents[1].text
    match = match.contents[3].contents[1]
    venue = match.find('span', class_='match-block__venue-name').text
    start_date = match.time['data-startdate']
    end_date = match.time['data-enddate']
    teams = match.find_all('span', class_='match-block__team-name')

    conclusion = conclusion.replace('\n', '')


    m = Match(teams[0].text, teams[1].text, start_date, end_date, venue, conclusion)
    data = np.append(data, m)
print(len(data))
```
